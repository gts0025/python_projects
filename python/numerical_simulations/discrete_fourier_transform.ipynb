```python
import numpy as np
import matplotlib.pyplot as plt

from fourier import*

time = np.linspace(0,10, 1000)
data = (abs(time - 5) < 2)*2
```

```python
plt.plot(time,data)
plt.title("stepping function")
```

```python




frequencies = np.array(forward(data, time,[0,1000],10000))
plt.plot(frequencies[:,0], frequencies[:,1])
#plt.plot(frequencies[:,0], frequencies[:,2])
plt.title("forward fourier series")
plt.ylabel("amplitude")
plt.xlabel("frequency")
plt.show()
```

```python
plt.plot(time,inverse(time,frequencies))
plt.title("inverse fourier series")
```

```python
derivative = fourier_derivative(data, time,[0,1000],10000)
plt.plot(time,derivative)
plt.title("fourier derivative")
plt.ylabel("f'(X)")
plt.xlabel("time")
```

```python

sample = np.array(forward(data, time,[600,650],100))
plt.plot(sample[:,0], sample[:,1])

plt.title("down sampled frequency")
plt.ylabel("amplitude")
plt.xlabel("frequency")
```

```python
plt.plot(time,inverse(time,sample))
plt.title("down sampled reconstruction")
```

```python

derivative = fourier_derivative(data, time,[600,650],100)
plt.plot(time,derivative)
plt.title("down sampled derivative")
plt.ylabel("amplitude")
plt.xlabel("frequency")
```
