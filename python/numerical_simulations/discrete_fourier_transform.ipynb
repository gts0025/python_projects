```python
%load_ext autoreload
%autoreload 2



import numpy as np
import matplotlib.pyplot as plt

from fourier import*

time = np.linspace(0,10, 1000)
data = (abs(time - 5) < 2)*10
```

```python
plt.plot(time,data)
plt.title("stepping function")
```

```python




frequencies = np.array(forward(data, time,[350,650],10000))
plt.plot(frequencies[:,0], frequencies[:,1])
#plt.plot(frequencies[:,0], frequencies[:,2])
plt.title("forward fourier series")
plt.ylabel("amplitude")
plt.xlabel("frequency")
plt.show()
```

```python
plt.plot(time,inverse(time,frequencies))
plt.title("inverse fourier series")
```

```python
derivative = fourier_derivative(time,frequencies)
 
 
plt.plot(time,derivative)
plt.title("fourier derivative")
plt.ylabel("f'(X)")
plt.xlabel("time")
```
