```python
import rasterio
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Replace with your GeoTIFF filename
espfilename = "esperantina_elevation.tif"

with rasterio.open(espfilename) as src:
    esp_elevation = src.read(1)          # Read the first band (e.g., elevation)
    esp_bounds = src.bounds              # Geographic extent
    esp_crs = src.crs                    # CRS info (WGS84/EPSG:4326)

batfilename = "batalha_elevation_data.tif"

with rasterio.open(batfilename) as src:
    bat_elevation = src.read(1)          # Read the first band (e.g., elevation)
    bat_bounds = src.bounds              # Geographic extent
    bat_crs = src.crs                    # CRS info (WGS84/EPSG:4326)

```

```python
plt.title("esperantina PI elevation data ")
plt.imshow(esp_elevation,cmap="jet")
plt.colorbar()
plt.show()
```

```python
plt.title("batalha PI elevation data ")
plt.imshow(bat_elevation,cmap="gray")

plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming bat_elevation is a 2D numpy array
shape = bat_elevation.shape
ny, nx = shape  # rows (y), cols (x)

x = np.arange(nx)   # columns
y = np.arange(ny)   # rows
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection="3d")

# Plot surface
ax.plot_surface(X, Y, bat_elevation, cmap="terrain", linewidth=0, antialiased=False)

ax.set_xlabel("X (pixel)")
ax.set_ylabel("Y (pixel)")
ax.set_zlabel("Elevation (m)")
plt.title("3D Terrain Surface")
plt.show()
```
