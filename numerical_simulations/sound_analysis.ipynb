```python

%load_ext autoreload
%autoreload

from fourier import *

import numpy as np 
import seaborn as sns
from scipy.io import wavfile
sps, data = wavfile.read("my_voice_record.wav")
data = np.array(data,np.int32)[:,0] 
```

```python
sns.lineplot(y = data ,x = np.linspace(0,data.shape[0]/sps,data.shape[0]))
```

```python
frequencies = forward(data,np.linspace(0,data.shape[0]/sps,data.shape[0]),[0,2000],4000)

sns.lineplot(x = frequencies[:,0], y = frequencies[:,1])
```

```python
cleaned = inverse(np.linspace(0,data.shape[0]/sps,data.shape[0]),frequencies)
sns.lineplot(y = cleaned ,x = np.linspace(0,cleaned.shape[0]/sps,cleaned.shape[0]))
```
